# Прогнозирование оттока клиента Банка

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

На основе данных из банка определить клиент, который может уйти.

```python
import pandas as pd
import numpy as np
from re import sub
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.model_selection import GridSearchCV
from random import randint
```

## Подготовка данных

### Импорт

```python
df = pd.read_csv('/datasets/Churn.csv')
df.info()
```


Признаки
* RowNumber — индекс строки в данных
* CustomerId — уникальный идентификатор клиента
* Surname — фамилия
* CreditScore — кредитный рейтинг
* Geography — страна проживания
* Gender — пол
* Age — возраст
* Tenure — сколько лет человек является клиентом банка
* Balance — баланс на счёте
* NumOfProducts — количество продуктов банка, используемых клиентом
* HasCrCard — наличие кредитной карты
* IsActiveMember — активность клиента
* EstimatedSalary — предполагаемая зарплата

Целевой признак
* Exited — факт ухода клиента

### Убираем бессмысленные признаки

```python
# Убираем признаки номера строки, фамилии и id клиента (от них решение не может быть зависеть)
df = df.drop(columns=['Surname', 'CustomerId', 'RowNumber'])
```

### Прямое кодирование

```python
# Преобразуем данные с учетом дамми ловушки
df_ohe =  pd.get_dummies(df, drop_first = True)
df_ohe.head()
```

### Приведение к правильному названию полей

```python
def snake_case(s):
  return '_'.join(
    sub('([A-Z][a-z]+)', r' \1',
    sub('([A-Z]+)', r' \1',
    s.replace('-', ' '))).split()).lower()

df_ohe.columns = [snake_case(c) for c in df_ohe.columns]
df_ohe.columns
```

###  Обрабатываем пустые значения

#### Cколько лет человек является клиентом банка

```python
# Cколько лет человек является клиентом банка - tenure
print('Всего пустых значений tenure:', df_ohe['tenure'].isna().sum(), 'из', len(df_ohe))
```

```python
# Список значений
df_ohe['tenure'].value_counts()
```

```python
# Заполняем случаными значениями от 1 до 10
def get_rnd_tenure():
    tenure = randint(1,10)
    return tenure

df_ohe.loc[df_ohe['tenure'].isnull(),'tenure'] = [randint(1,10) for i in range(0,df_ohe['tenure'].isna().sum())]
print('Всего пустых значений tenure:', df_ohe['tenure'].isna().sum(), 'из', len(df_ohe))
```

```python
# Список значений после замены
df_ohe['tenure'].value_counts()
```

### Разбивка данных

```python
# Разбивение выборки на обучающую и тестовую
target = df_ohe['exited']
features = df_ohe.drop('exited', axis=1)

# Разбиваем на тестовую и обучающую выборку
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.20, random_state=12345)
# Разбиваем обучающую выборку на тренировочную и валидационную
features_train, features_valid, target_train, target_valid = train_test_split(features_train, target_train, test_size=0.25, random_state=12345)
print(features_train.shape)
print(features_valid.shape)
print(features_test.shape)
```

### Стандартизация данных

```python
# Стандартизуем данные
pd.options.mode.chained_assignment = None

numeric = ['credit_score', 'age', 'tenure', 'balance', 'num_of_products', 'has_cr_card','is_active_member', 'estimated_salary']

scaler = StandardScaler()
scaler.fit(features_train[numeric])

features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])

features_train.head()
```

### Модель логистической регрессии

```python
# Создаем модель
model = LogisticRegression(random_state=12345, solver='liblinear')
model.fit(features_train, target_train) 

# Находим показатель
predicted_train = model.predict(features_train)
print("F1 на тренировочной выборке:", f1_score(target_train, predicted_train))

predicted_valid = model.predict(features_valid)
print("F1 на валидационной выборке:", f1_score(target_valid, predicted_valid))

predicted_test = model.predict(features_test)
print("F1 на тестовой выборке:", f1_score(target_test, predicted_test))

#Находим площадь под ROC -кривой — AUC-ROC
probabilities_train = model.predict_proba(features_train)
probabilities_one_train = probabilities_train[:, 1]
auc_roc_train = roc_auc_score(target_train, probabilities_one_train)
print('auc_roc на тренировочной выборке:', auc_roc_train)

probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
auc_roc_valid = roc_auc_score(target_valid, probabilities_one_valid)
print('auc_roc на валидационой выборке:', auc_roc_valid)

probabilities_test = model.predict_proba(features_test)
probabilities_one_test = probabilities_test[:, 1]
auc_roc_test = roc_auc_score(target_test, probabilities_one_test)
print('auc_roc на тестовой выборке:', auc_roc_test)
```

### Модель случайного леса

```python
# Поиск оптимальных гиперпараметров случайного леса GridSearch
parametrs = { 'n_estimators': range (100, 151, 10),
              'max_depth': range (1,16, 2),
              'min_samples_leaf': range (1,8),
              'min_samples_split': range (2,10,2) }

clf = RandomForestClassifier()
grid = GridSearchCV(clf, parametrs, cv=5, scoring='f1', verbose=10)
grid.fit(features_train, target_train)

grid.best_params_
```

```python
# Оптимальные параметры
best_params ={'max_depth': 15, 'min_samples_leaf': 2, 'min_samples_split': 2, 'n_estimators': 140}
```

```python
# Создаем модель
model = RandomForestClassifier(random_state=12345, n_estimators=best_params['n_estimators'],
                               max_depth=best_params['max_depth'], 
                               min_samples_leaf=best_params['min_samples_leaf'],
                               min_samples_split=best_params['min_samples_split'])
# Обучаем на тренировочных данных
model.fit(features_train, target_train)

print('F1 наилучшей модели на тренировочной выборке:', f1_score(target_train, model.predict(features_train)))
print("F1 наилучшей модели на валидационной выборке:", f1_score(target_valid, model.predict(features_valid)))
print('F1 наилучшей модели на тестовой выборке:', f1_score(target_test, model.predict(features_test)))


#Находим площадь под ROC -кривой — AUC-ROC
probabilities_train = model.predict_proba(features_train)
probabilities_one_train = probabilities_train[:, 1]
auc_roc_train = roc_auc_score(target_train, probabilities_one_train)
print('auc_roc на тренировочной выборке:', auc_roc_train)

probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
auc_roc_valid = roc_auc_score(target_valid, probabilities_one_valid)
print('auc_roc на валидационой выборке:', auc_roc_valid)
                
probabilities_test = model.predict_proba(features_test)
probabilities_one_test = probabilities_test[:, 1]
auc_roc_test = roc_auc_score(target_test, probabilities_one_test)
print('auc_roc на тестовой выборке:', auc_roc_test)
```

### Выводы

* Целевой признак имеет дисбаланс в отношении 5:1 (положительных в 5 раз меньше). Имеет смысл применять техники борьбы с дисбалансом

### Метод увеличения выборки

```python
# Коэффициент дисбаланса
balance_coef = int(round(1 /(len(df_ohe[df_ohe['exited'] == 1])/ len(df_ohe['exited'])), 0))
balance_coef
```

```python
# Применяем технику upsample
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled


features_upsampled, target_upsampled = upsample(features_train, target_train, balance_coef)
features_upsampled.shape
```

```python
# Проверяем баланс целевого признака
target_upsampled.value_counts().plot(kind='pie');
```

####  Модель логистической регрессии

```python
# Создаем модель
model = LogisticRegression(random_state=12345, solver='liblinear')
model.fit(features_upsampled, target_upsampled) 

# Находим показатель
predicted_train = model.predict(features_upsampled)
print("F1 на тренировочной выборке:", f1_score(target_upsampled, predicted_train))

predicted_valid = model.predict(features_valid)
print("F1 на валидационной выборке:", f1_score(target_valid, predicted_valid))

predicted_test = model.predict(features_test)
print("F1 на тестовой выборке:", f1_score(target_test, predicted_test))

#Находим площадь под ROC -кривой — AUC-ROC
probabilities_train = model.predict_proba(features_upsampled)
probabilities_one_train = probabilities_train[:, 1]
auc_roc_train = roc_auc_score(target_upsampled, probabilities_one_train)
print('auc_roc на тренировочной выборке:', auc_roc_train)

probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
auc_roc_valid = roc_auc_score(target_valid, probabilities_one_valid)
print('auc_roc на валидационой выборке:', auc_roc_valid)

probabilities_test = model.predict_proba(features_test)
probabilities_one_test = probabilities_test[:, 1]
auc_roc_test = roc_auc_score(target_test, probabilities_one_test)
print('auc_roc на тестовой выборке:', auc_roc_test)
```

### Порог классификации


#### Модель логистической регрессии

```python
model = LogisticRegression(random_state=12345, solver='liblinear')
model.fit(features_train, target_train)
probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

best_threshold = float("-inf")
best_score = float("-inf")
best_test = float("-inf")
for threshold in np.arange(0, 0.5, 0.02):
    predicted_valid = probabilities_one_valid > threshold
    score_valid = f1_score(target_valid, predicted_valid)
    
    predicted_test = probabilities_one_test > threshold
    score_test = f1_score(target_test, predicted_test)
    if score_valid > best_score:
        best_threshold = threshold
        best_score = score_valid
        best_test = score_test


print("Лучшие показатели валидационнная выборка: Порог = {:.2f} | F1 = {:.4f}".format(best_threshold, best_score))
print("Лучшие показатели тестовая  выборка: Порог = {:.2f} | F1 = {:.4f}".format(best_threshold, best_test))
```

#### Модель случайного леса


```python
# Создаем модель
model = RandomForestClassifier(random_state=12345, n_estimators=best_params['n_estimators'],
                               max_depth=best_params['max_depth'], 
                               min_samples_leaf=best_params['min_samples_leaf'],
                               min_samples_split=best_params['min_samples_split'])

model.fit(features_train, target_train)
probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

best_threshold = float("-inf")
best_score = float("-inf")
for threshold in np.arange(0, 0.5, 0.02):
    predicted_valid = probabilities_one_valid > threshold
    
    score_valid = f1_score(target_valid, predicted_valid)
    if score_valid > best_score:
        best_threshold = threshold
        best_score = score_valid


print("Лучшие показатели на валидацинной выборке: Порог = {:.2f} | F1 = {:.4f}".format(best_threshold, best_score))

predicted_test = probabilities_one_test > best_threshold
predicted_test = probabilities_one_test > threshold
print('F1 наилучшей модели на тестовой выборке:', f1_score(target_test, predicted_test))
```

### Выводы

* Применили две техники борьбы с дисбалансом в выборке: взвешивание классов в параметрах модели и метод увеличения выборки 
* При дисбалансе данных между положительными и отрицательными значениями необходимо использовать F1-оценку, т.к.ROC усредняет пороговые значения. Чтобы показатель F1 был высоким, точность и полнота должны быть высокими.
* После применения техник в большинстве  модели улучшились (показатели для тестовой выборки):

| Модель                  | Без баланса F1 | Взввешивание классов | Увеличение выборки | Порог классификации |
| ----------------------- | -----------    | -------------------- | ------------------ | ------------------  |
| Линейная регрессия      | 0.3047         | 0.5041               |  0.4885            | 0.5946              |
| Случайный лес           | 0.5865         | **0.6208**           |  **0.6335**        | **0.6321**              |

* Целевой показатель > 0.59 достигнут на тестовой выборке случайного леса после применения техник балансировки классов
* Лучший показатель у модели случайного леса после применения техни
