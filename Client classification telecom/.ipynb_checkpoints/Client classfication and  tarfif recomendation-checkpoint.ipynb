# Классификация клиентов телеком компании и рекомендация тарифа

На основе данных предложить клиенту лучший тариф.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
import matplotlib.pyplot as plt
import numpy as np
```

## Импорт данных

```python
df = pd.read_csv("/datasets/users_behavior.csv")
df.info()
```

Каждый объект в наборе данных — это информация о поведении одного пользователя за месяц. Известно:
* сalls — количество звонков,
* minutes — суммарная длительность звонков в минутах,
* messages — количество sms-сообщений,
* mb_used — израсходованный интернет-трафик в Мб,
* is_ultra — каким тарифом пользовался в течение месяца («Ультра» — 1, «Смарт» — 0).

## Разбивка данных на выборки

```python
# random_state - для повторяемости результата
random_state = 123456
# Разбиваем на тестовую и обущающую выборку 80/20
df_learn, df_test = train_test_split(df, test_size=0.20, random_state=random_state)
# Разбиваю обучающую выборку на собственно обучающую и валидационную
df_train, df_valid = train_test_split(df_learn, test_size=0.20, random_state=random_state)

# Выделяем фичи и целевой признак для обучающей выборки
features_train = df_train.drop(['is_ultra'], axis = 1)
target_train = df_train['is_ultra']

# Выделяем фичи и признак для валидационной выборки
features_valid = df_valid.drop(['is_ultra'], axis = 1)
target_valid = df_valid['is_ultra']

# Выделяем фичи и признак для тестовой выборки
features_test = df_test.drop(['is_ultra'], axis = 1)
target_test = df_test['is_ultra']
```

## Исследование модели

### Анализ фич

```python
# Анализируеум взаимосвязи (корреляцию) фич
df.corr()
```

```python
# Строим гистрограммы распределения 
grr = pd.plotting.scatter_matrix(df, c=df['is_ultra'], figsize=(10,10), marker='o',hist_kwds={'bins': 20}, s=40, alpha=.6);
```

* Есть прямая положительная корреляция количества минут и количетсво звонов. Т.о, теоретически, можем использовать только один из этих признаков
* Пользователи тарифа ultra в общем тратять большее количество минут, смс и Гбт

### Модель решающего дерева

```python
best_model_dtc = None
best_result_dtc = 0
for max_depth in range(1, 11):
    # Создаем модель
    model = DecisionTreeClassifier(random_state = random_state, max_depth = 3) 
    # Обучаем модель
    model.fit(features_train, target_train);
    # Посчитаем качество модели на валидационной выборке
    result = model.score(features_valid, target_valid)
    if result > best_result_dtc:
        best_model_dtc = model
        best_result_dtc = result

print("Accuracy наилучшей модели на валидационной выборке:", best_result_dtc)
print("Параметры наилучшей модели:", best_model_dtc)
```

### Модель случайного леса

```python
best_model_rtc = None
best_result_rtc = 0
for est in range(1, 11):
    # Создаем модель
    model = RandomForestClassifier(random_state = random_state, n_estimators = 3) 
    # Обучаем модель
    model.fit(features_train, target_train);
    # Посчитаем качество модели на валидационной выборке
    result = model.score(features_valid, target_valid)
    if result > best_result_rtc:
        best_model_rtc = model
        best_result_rtc = result

print("Accuracy наилучшей модели на валидационной выборке:", best_result_rtc)
print("Параметры наилучшей модели:", best_model_rtc)
```

### Модель логистической регрессии

```python
# Создаем модель
model_lr = LogisticRegression(random_state = random_state) 
# Обучаем модель
model_lr.fit(features_train, target_train);
# Посчитаем качество модели на валидационной выборке
result = model_lr.score(features_valid, target_valid)

print("Accuracy модели на валидационной выборке:", result)
```

## Проверьте модель на тестовой выборке

### Проверям модель решающего дерева

```python
result = best_model_dtc.score(features_test, target_test)
print('Accuracy модели на тестовой выборке:', result)
```

### Проверям модель случайного леса

```python
result = best_model_rtc.score(features_test, target_test)
print('Accuracy модели на тестовой выборке:', result)
```

### Проверям модель логистической регрессии

```python
result = model_lr.score(features_test, target_test)
print('Accuracy модели на тестовой выборке:', result)
```

## Проверка модели на адекватность

### Проверка Accuracy собственной реализацией на примере решающего леса

```python
y_true = list(target_test)
y_pred = best_model_dtc.predict(features_test)

counter = 0
for i in range(0, len(y_true)):
   if y_true[i] == y_pred[i]:
       counter+=1

result = counter / len(y_true)
print('Accuracy модели на тестовой выборке:', result)
```

### Проверка точности модели
